## My Pandassql engine doesn't support cast and convert Functions so i must convert visit_date in order to avoid error fields ##

```python
r_visitors['visit_date'] = r_visitors['visit_datetime'].str[6:10]+'-'+r_visitors['visit_datetime'].str[3:5]+'-'+r_visitors['visit_datetime'].str[0:2]
r_visitors['visit_date'] = pd.to_datetime(r_visitors['visit_date'])
r_visitors['visit_date'] = r_visitors['visit_date'].dt.date
r_visitors['reserve_visitors'] = r_visitors['reserve_visitors'].astype(float)
```

```python
def pysqldf(q):
    "Función para generar query globalmente"
    return sqldf(q, globals())
```

# Answer 1:

```python
q="""

SELECT 
    a.id
    ,c.genre_name
    ,c.area_name
    ,AVG(a.reserve_visitors) as average_visitors
    
FROM r_visitors a 

LEFT OUTER JOIN date_info b ON a.visit_date = b.calendar_date
LEFT OUTER JOIN store_info c ON a.id = c.store_id

WHERE b.holiday_flg = 1

GROUP BY 1,2,3
ORDER BY average_visitors DESC;


"""
```

```python
pysqldf(q)
```

# Answer 2

```python
q="""

SELECT 
     b.day_of_week
    ,AVG(a.reserve_visitors) as average_visitors
    
FROM r_visitors a 

LEFT OUTER JOIN date_info b ON a.visit_date = b.calendar_date
LEFT OUTER JOIN store_info c ON a.id = c.store_id

GROUP BY 1
ORDER BY average_visitors DESC;


"""
```

```python
pysqldf(q)
```

```python
# Friday is the top day for average visitors
```

# Answer 3

```python
q="""

SELECT 
     MAX(a.visit_date)
     ,b.day_of_week

FROM r_visitors a 

LEFT OUTER JOIN date_info b ON a.visit_date = b.calendar_date;

"""
```

```python
pysqldf(q)
```

```python
q="""

SELECT 
     
     CASE
         WHEN a.visit_date BETWEEN '2017-05-04' AND '2017-05-10' THEN 'Week 1'
         WHEN a.visit_date BETWEEN '2017-05-11' AND '2017-05-17' THEN 'Week 2'
         WHEN a.visit_date BETWEEN '2017-05-18' AND '2017-05-24' THEN 'Week 3'
         WHEN a.visit_date BETWEEN '2017-05-25' AND '2017-05-31' THEN 'Week 4'
         Else 'NA'
         END AS Week_Number
     
     ,SUM(a.reserve_visitors) as total_visitors
     ,LAG(sum(a.reserve_visitors),1,0) over() as Movement
     ,SUM(a.reserve_visitors) / LAG(sum(a.reserve_visitors),1,0) over() -1 as Growth_WoW
     
     
         
FROM r_visitors a 

WHERE Week_Number NOT IN('NA')

GROUP BY 1;

"""
```

```python
pysqldf(q)
```
